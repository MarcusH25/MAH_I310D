## What are `repositories`, `git` and `GitHub`?

**Repository** : A place where your code is stored. A fancier term for directory.

**GitHub** : A server (online / cloud storage), that publishes your projects online for others to download and build off of. Once others are done with their work, they upload their changes back to GitHub

**git**: A program that runs on your local machine and tracks all the changes you or others have been making in the repository since the start of the project. It also allows you to download or upload projects from a server like GitHub.

## 4. Now let's work on our project

- Open area_of_circle.py in your favorite text editor (notepad / TextEdit etc). 
- Make some changes to your file. Save changes. 
- Come back to Terminal and go inside your cloned directory using `cd <path_to_your_directory>` comand
- Execute the following command to check the change status

    `git status`
    
    **(What do you see?)**

- Execute the following command to see the changes:

    `git diff`


- Execute the following command for UPDATING your new changes:

    `git add -u`


- Execute the following command for committing your changes (Add a message to your commit):

    `git commit -m "


- Execute the following command for pushing your changes to github

    `git push -u origin master`
    
 
Now, every time you make changes to your files, you follow the above steps to make sure your changes are tracked and reflected in the server.
